```python
import pandas as pd

url = "https://raw.githubusercontent.com/gmauricio-toledo/NLP-LCC/main/Rest-Mex/Rest-Mex_2025_test.xlsx"
test_df = pd.read_excel(url)
test_df.drop(columns='ID',inplace=True)
test_size = test_df.shape[0]
test_df.head()
```

## Preprocesamiento

```python
def check_data(df,column):
    # Viendo si hay valores nulos
    print(f"nulos: \n{df.isnull().sum()}")
    # Revisando valores unicos de sentiment
    print("---------")
    #viendo la longitud de string mas pequeña en review
    print(df[column].str.len().min())
    print("---------")
    #viendo la mas larga
    print(df[column].str.len().max())
    print("---------")
    #Viendo el promedio
    print(df[column].str.len().mean())
    print("---------")
    #viendo valores unicos para review
    print(f"Valores repetidos: {df.duplicated().sum()}")
    print("---------")
```

```python
check_data(test_df,'Review')
```

```python
test_df["title_and_text"] = test_df['Title'].fillna("").astype(str) + " " + test_df["Review"].astype(str)
test_df = test_df.drop(columns=['Title','Review'], axis=1)
#test_df.drop_duplicates(subset=['title_and_text'], inplace=True)
test_df.head()
```

```python
check_data(test_df,'title_and_text')
```

```python
import spacy
from spacy.lang.es.stop_words import STOP_WORDS
import nltk
import string
import re
# Cargar el modelo de spaCy en español
nlp = spacy.load("es_core_news_sm")

# Quitando StopWords
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = STOP_WORDS
```

```python
# Quitando StopWords
remove = ["no", "mal", "bien", "buen", "mejor", "peor", "hay"] #si, tambien, mucho, poco, mas, menos, bastante, muy, nada
for word in remove:
    if word in stop_words:
        stop_words.remove(word)
```

```python
def preprocess(df):
    #Minusculas
    df = df.str.lower()
    #Quitar simbolos
    df = df.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
    #Quitar stopwords
    df = df = df.apply(lambda x: re.sub(r"[^a-zA-Z0-9 ]", "", x))
    #quitar signos de puntuacion
    df = df.str.translate(str.maketrans('', '', string.punctuation))

    #Lematizar:
    return df
```

```python
preprocessed_test = test_df.copy()
preprocessed_test.head()
```

```python
preprocessed_test = preprocess(preprocessed_test['title_and_text'].astype(str))
preprocessed_test.head()
```

```python
import pickle
# Importamos el vectorizador que usamos para entrenar el modelo
with open('tfidf_vectorizer.pkl', 'rb') as f:
    vectorizer = pickle.load(f)
```

```python
#Vectorizamos con tf-idf
preprocessed_test = vectorizer.transform(preprocessed_test)
```

```python
#Importamos el modelo
import pickle
with open('NB_model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
predicciones = model.predict(preprocessed_test)
```
