```python
texto = df_tag_0["text"][2]
```

```python
import nltk
from nltk import bigrams, trigrams
from nltk.corpus import reuters
from collections import defaultdict
from string import punctuation

nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
stopword_es = nltk.corpus.stopwords.words('spanish')
stopword_es
```

```python
punctuations = list(punctuation)

words = [t for t in nltk.word_tokenize(texto) if t not in punctuations and t not in stopword_es]

words
```

```python
from collections import Counter

frequent_words = Counter(words)
frequent_words.most_common(10)
```

```python
df_tag_0.reset_index()
df_tag_0["text"]
```

```python
df_tag_1 = df_tag_1["text"].reset_index()
```

```python
def get_all_text(df):
  list_of_tokens = []
  for i in range(0,df["text"].size):
    texto = df["text"][i]
    words = [t.lower() for t in nltk.word_tokenize(texto) if t not in punctuations and t not in stopword_es]
    list_of_tokens.extend(words)
  return list_of_tokens
```

```python
all_text_tokens = get_all_text(df_tag_0)
```

```python
all_text_freq = Counter(all_text_tokens)
all_text_freq.most_common(10)
```

```python
all_text_tokens = get_all_text(df_tag_1)
all_text_freq = Counter(all_text_tokens)
all_text_freq.most_common(10)
```
